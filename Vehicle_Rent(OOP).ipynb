```python
import time
class Vehicle:
    def __init__(self):
        self.class_name='Vehicle'
        self.menu=' Main Menu '
        self.var={'A':'Car Menu','B':'Bike Menu','Q':'Quit'}
        
    def print_menu(self):
        print('\n{:*^20s}'.format(self.menu))
        for i,j in self.var.items():
            print(i+'.',j)
            time.sleep(0.7)
            
    def count_good(self):
        print(self.count_of_good)
        
    def buy_return_good(self,s,k=False):
        if k==True:
            self.count_of_good +=s
        else:    
            self.count_of_good -= s
                     
            
```

```python
class Car(Vehicle):
    
    def __init__(self):
        self.class_name='Car'
        self.count_of_good=10
        self.menu=' Car Menu '
        self.var= {'1':'Display available cars',
                   '2':'Request a car on hourly basis $ 10',
                   '3':'Request a car on daily basis $ 192',
                   '4':'Return a car',
                   '5':'Main Menu',
                   '6':'Exit'
                  }
        
```

```python
class Bike(Vehicle):
    
    def __init__(self):
        self.class_name='Bike'
        self.count_of_good=100
        self.menu=' Bike Menu '
        self.var ={'1':'Display available bikes',
                   '2':'Request a bike on hourly basis $ 5',
                   '3':'Request a bike on daily basis $ 84',
                   '4':'Return a bike',
                   '5':'Main Menu',
                   '6':'Exit'
                  }
```

```python
class Consumer(Car,Bike):
   
    def obtain(self,a):
        if a=='A':
            Car.__init__(self)
        elif a=='B':    
            Bike.__init__(self)
        
    def functions(self,count):
        
        try:
             int(count)
        except ValueError:
             self.obtain(count)
        else:    
            if self.var[count]=='Display available cars' or self.var[count]=='Display available bikes':
                 self.count_good()
            
            elif (self.var[count]=='Request a car on hourly basis $ 10' or self.var[count]=='Request a bike on hourly basis $ 5') or  (self.var[count]== 'Request a bike on daily basis $ 84' or self.var[count]=='Request a car on daily basis $ 192'):                                                                                                                             
                s=int(input('how much: '))
                self.buy_return_good(s)     
                
            elif  self.var[count]=='Return a car' or self.var[count]=='Return a bike':   
                s=int(input('how much: '))
                self.buy_return_good(s,True)
            elif self.var[count]=='Main Menu':
                Vehicle.__init__(self)
            
```

```python
obj=Vehicle()
obj.print_menu()
obj=Consumer()
x=input('enter: ').upper()
obj.obtain(x)
while True:
    if x=='Q':
        break
    while True:
        
        obj.print_menu()
        x=input('enter2: ').upper()
        if ( (obj.class_name==('Car') or obj.class_name==('Bike'))  and  x=='6') or (obj.class_name=='Vehicle' and x=='Q') :
            break   
        obj.functions(x)
    break
```
